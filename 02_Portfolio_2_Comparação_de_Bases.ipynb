# Preparação, organização e estruturação dos dados.

### Extração das duas bases

```python
# CARREGAR O DATASET em formato Excel, tipo de arquivo .XLSX.

# O arquivo na origem possui mais de uma planilha, por isso foi acrescentado um direcionamento para a Lista1.

pb = pd.read_excel('/content/drive/MyDrive/PYT/1_Portfolio/0_Estudo_Plan_Produtos/Prod1b.xlsx', sheet_name='Sheet1')

pc = pd.read_excel('/content/drive/MyDrive/PYT/1_Portfolio/0_Estudo_Plan_Produtos/Prod1c.xlsx', sheet_name='Sheet1')
```

### Apresentação do DATASET: Produtos 1

Produto (SKU), Título, Marca, Categoria, Valor de Custo, Preço de Venda, Estoques, Armazém, Filial e Tabela.


```python
# Primeiros registros base A = pb

pb.head(21)
```

```python
# Primeiros registros base B = pc

pc.head(21)
```

```python
# Unir os dataframes usando o SKU como chave de junção
df_merged = pd.merge(pb, pc, on='SKU', suffixes=('_a', '_b'))

df_merged.head()
```

```python
# Comparar os preços e identificar discrepâncias
df_merged['discrepancia'] = df_merged['PRECO_VISTA_a'] - df_merged['PRECO_VISTA_b']

df_merged.head()
```

```python
# Visualizar as discrepâncias
discrepancias = df_merged[df_merged['discrepancia'] != 0]
print(discrepancias)
```

```python
# Selecionar apenas as colunas necessárias

discrepanciasz = discrepancias[['SKU', 'TITULO_a', 'PRECO_VISTA_a', 'PRECO_VISTA_b', 'discrepancia']]

print(discrepanciasz)
```

```python
# Renomear os campos para melhor legibilidade
discrepancias2 = discrepanciasz.rename(columns={'TITULO_a': 'Título', 'PRECO_VISTA_a': 'Preço A', 'PRECO_VISTA_b': 'Preço B'})

discrepancias2.head(10)
```

Resultado: Análise_1:
No relatório obtido temos a coluna "discrepancia" contendo o valor da diferença entre Preço A e Preço B.

Ao final gerei o relatório Discrepancia em formatos de Excel e CSV para a área comercial da empresa EX-Vendas.

```python
# Gerando arquivo em formato XLSX, Excel:

discrepancias2.to_excel('/content/drive/MyDrive/PYT/1_Portfolio/0_Estudo_Plan_Produtos/Discrepancia.xlsx', index=False)
```

```python
# Gerando arquivo em formato CSV:

discrepancias2.to_csv('/content/drive/MyDrive/PYT/1_Portfolio/0_Estudo_Plan_Produtos/Discrepancia.CSV', index=False)
```
